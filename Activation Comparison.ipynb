```python
import pickle
with open("activation_steepening.pickle", "r") as datafile:
    raw = pickle.load(datafile)
```

```python
import matplotlib.pyplot as plt
```

```python
acc_2d = [item[0]["acc"] for item in raw[4]]
acc_1d = [item for sublist in acc_2d for item in sublist]
val_acc_2d = [item[0]["val_acc"] for item in raw[4]]
val_acc_1d = [item for sublist in val_acc_2d for item in sublist]

acc_lens = [len(item) for item in acc_2d]
acc_cum = [sum(acc_lens[0:i+1]) for i in range(len(acc_lens))]

plt.plot(acc_1d)
plt.plot(val_acc_1d)

plt.plot(acc_cum, [acc_1d[i-1] for i in acc_cum], "ro")
```

```python
acc_2d = [[item[0]["acc"] for item in trial] for trial in raw]
acc_1d = [[item for sublist in trial for item in sublist] for trial in acc_2d]

val_acc_2d = [[item[0]["val_acc"] for item in trial] for trial in raw]
val_acc_1d = [[item for sublist in trial for item in sublist] for trial in val_acc_2d]

acc_lens = [[len(item) for item in trial] for trial in acc_2d]
acc_cum = [[sum(trial[0:i+1]) for i in range(len(trial))] for trial in acc_lens]

for acc, val_acc, cum, color in zip(acc_1d, val_acc_1d, acc_cum, ["b", "r", "g", "r", "c"]):
    plt.plot(acc)#, fmt=color)
    #plt.plot(val_acc)#, fmt=color)
    
    plt.plot(cum, [acc[i-1] for i in cum], color + "o")
```
